```python
X = df[['Luas Wilayah','Jumlah']].values
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
x = scaler.fit_transform(X)
```

```python
x
```

```python
scaled_df = pd.DataFrame(x, columns=['Luas Wilayah', 'Jumlah Penduduk'])
print(scaled_df)
```

```python
from sklearn.cluster import KMeans
from sklearn.metrics import davies_bouldin_score
from sklearn.utils.validation import check_array

for n_clusters in range(2, 11):
    # Melakukan klastering menggunakan K-Means
    kmeans = KMeans(n_clusters=n_clusters, random_state=42)
    labels = kmeans.fit_predict(x)

    # Menghitung Davies-Bouldin Index
    dbi_score = davies_bouldin_score(check_array(x), labels)

    print(f'Cluster: {n_clusters}, Davies-Bouldin Index: {dbi_score}')
```

```python
from sklearn.cluster import KMeans
from sklearn.metrics import davies_bouldin_score
from sklearn.utils.validation import check_array

for n_clusters in range(2, 11):
    # Melakukan klastering menggunakan K-Means
    kmeans = KMeans(n_clusters=n_clusters, random_state=42)
    labels = kmeans.fit_predict(x)

    # Menghitung Davies-Bouldin Index
    dbi_score = davies_bouldin_score(check_array(x), labels)

    print(f'Cluster: {n_clusters}, Davies-Bouldin Index: {dbi_score}')
```

```python
from sklearn.metrics import silhouette_score

silhouette_avg = silhouette_score(x, kmeans.labels_)
print("Silhouette Coefficient:", silhouette_avg)
```

```python
from sklearn.metrics import silhouette_score
from sklearn.cluster import KMeans

for n_cluster in range(2,11):
    kmeans = KMeans(n_clusters=n_cluster). fit(x)
    label=kmeans.labels_
    silhouette_avg = silhouette_score(x, label, metric ='euclidean')
    print("For n_cluster={}, The Silhouette Score {}".format(n_cluster, silhouette_avg))
```

```python
from sklearn.cluster import KMeans
cs = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters = i)
    kmeans.fit(x)
    cs.append(kmeans.inertia_)
cs
```

```python
plt.plot(range(1, 11), cs)
plt.title('The Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('CS')
plt.show()
```
